```python
#Delete first few columns, excul,leave thsoe 'useful' variables
colnames = df.columns

df_onlyvar = df.drop(columns = colnames[0])
```

```python
#Here, since there are 75 columns (drop the first column), divide those variables

part1 = list({74}) + list(range(0,25))
part2 = list({74}) + list(range(26,50))
part3 = list({74}) + list(range(51,73))

df_onlyvar_1 = df_onlyvar[df_onlyvar.columns[part1]]
df_onlyvar_2 = df_onlyvar[df_onlyvar.columns[part2]]
df_onlyvar_3 = df_onlyvar[df_onlyvar.columns[part3]]
```

```python
#Make Correlation Heatmap
correlation_p1 = df_onlyvar_1.corr()
correlation_p2 = df_onlyvar_2.corr()
correlation_p3 = df_onlyvar_3.corr()

axis_corr = sns.heatmap(
correlation_p1,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),
square=True
)

plt.show()

axis_corr = sns.heatmap(
correlation_p2,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),
square=True
)

plt.show()

axis_corr = sns.heatmap(
correlation_p3,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),
square=True
)

plt.show()
```
